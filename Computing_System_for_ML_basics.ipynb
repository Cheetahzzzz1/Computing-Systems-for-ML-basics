```python
# Function to calculate the mean for each dimension
def findMean(points):
    total_points = len(points)  # Total number of points that we have
    num_dimensions = len(points[0])  # This will assume that all points have the same dimension.

    # Initialize a list to store mean for each dimension
    mean = [0] * num_dimensions

    for point in points: # Iterates over all the points
        for i in range(num_dimensions): # Iterates over dimension of current points
            mean[i] += point[i]

    for i in range(num_dimensions): # This will tell us the mean by dividing the sum by total number of points
        mean[i] /= total_points

    return mean

# Function to calculate the standard deviation for each dimension
def findStd(points):
    total_points = len(points)  # Total number of points
    num_dimensions = len(points[0])  # Number of dimensions per point
    mean = findMean(points)  # Compute the mean for each dimension

    # Initialize a list to store variance for each dimension
    variance = [0] * num_dimensions

    for point in points: # Iterates over each point in the list
        for i in range(num_dimensions): # Iterates over every dimension.
            variance[i] += (point[i] - mean[i]) ** 2 # This is the squared sum of difference between each point and the calculated mean.

    for i in range(num_dimensions): # This will calculate the variance
        variance[i] /= total_points

    std_dev = [variance[i] ** 0.5 for i in range(num_dimensions)] # This will give us Standard Deviation which is a square root of variance.

    return std_dev

# Displaying the output with the given points

points = [(2,2,2), (3,3,3), (4,4,4), (6,0,0), (0,6,0), (0,0,6)]

print("Mean:",findMean(points))
print("Standard Deviation:", findStd(points))
```

```python
def InnerProduct(vector1, vector2):

    if len(vector1) != len(vector2): # We must first ensure that the vectors are of same length
        print("Error: Vectors must have the same length.")
        return

    dot_product = 0 # It will store element wise product of the two vectors

    for i in range(len(vector1)): # This will iterate through every index
        dot_product += vector1[i] * vector2[i] # Multiply corresponding elements of both vectors and add to the dot product

    return dot_product # This will return the dot product.

# When the vectors are of same length
vector1 = [24, 34, 56]  # First vector
vector2 = [19, 32, 48]  # Second vector

print("Inner Product:", InnerProduct(vector1, vector2))

# When vectors are of different length

vector3 = [38, 46, 52, 21]
vector4 = [15, 14]

print("Inner Product:", InnerProduct(vector3, vector4))
```

```python
# Define a function to read the file and process the data
def process_baby_names(filename):

    boy_names = {} # Dictionary to store boy names by year
    girl_names = {} # Dictionary to store girl names by year
    all_names = {} # Dictionary to store all the names

    # Open the file for reading
    file = open(filename, 'r')

    # Read the first line (header) and ignore it
    file.readline()

    # Read each line in the file
    while True:
        line = file.readline()
        if not line:  # Break if end of file is reached
            break

        # Split the line into components (year, name, percent, sex)
        components = [] # List to store split values
        current_component = ""
        in_quotes = False # Flag to check if inside is a quoted name

        for char in line: # Looping through each character in the line
            if char == '"':  # Toggle quotes
                in_quotes = not in_quotes
            elif char == ',' and not in_quotes:  # Split on comma if not within quotes
                components.append(current_component)
                current_component = ""
            else:
                current_component += char

        components.append(current_component)  # Add the last component

        year = int(components[0]) # Here we are converting year to an integer
        name = components[1].strip('"') # We are removing quotes from the names
        percent = float(components[2]) # Converting percentage to a float
        sex = components[3].strip('"') # Removing quotes from gender

        # Store names based on gender and year
        if sex == 'boy':
            if year not in boy_names:
                boy_names[year] = {} # Initializing a dictionary for the year
            if name not in boy_names[year]:
                boy_names[year][name] = 0.0
            boy_names[year][name] += percent # Adding percentage to the name
        else:
            if year not in girl_names:
                girl_names[year] = {}
            if name not in girl_names[year]:
                girl_names[year][name] = 0.0
            girl_names[year][name] += percent

        # Store total percentages for all names
        if name not in all_names:
            all_names[name] = 0.0
        all_names[name] += percent

    file.close()  # Close the file

    # Find the most popular boy name for each year
    popular_boy_names = {} # Dictionary to store most popular boy names
    for year in boy_names: # Looping through each year
        max_name = "" # This will store the name with the highest percentage
        max_percent = -1.0 # This will store the highest percentage
        for name in boy_names[year]:
            if boy_names[year][name] > max_percent:
                max_percent = boy_names[year][name] # Here we are updating maximum percentage
                max_name = name
        popular_boy_names[year] = max_name # Used to store the name of the most popular boy of a [articular year]

    # Find the most popular girl name for each year
    popular_girl_names = {}                       # Everything is same as above but for girl names
    for year in girl_names:
        max_name = ""
        max_percent = -1.0
        for name in girl_names[year]:
            if girl_names[year][name] > max_percent:
                max_percent = girl_names[year][name]
                max_name = name
        popular_girl_names[year] = max_name

    # Find the most popular name across all years
    most_popular_name = ""                        # Everything is same as above but for all names
    highest_percent = -1.0
    for name in all_names:
        if all_names[name] > highest_percent:
            highest_percent = all_names[name]
            most_popular_name = name

    return popular_boy_names, popular_girl_names, most_popular_name

# Print results for most popular boy names
print("(a) Most popular boy names by year:")
for year in sorted(popular_boy_names.keys()):
    print(f"{year}: {popular_boy_names[year]}")

# Print results for most popular girl names
print("\n(b) Most popular girl names by year:")
for year in sorted(popular_girl_names.keys()):
    print(f"{year}: {popular_girl_names[year]}")

# Print result for most popular name overall
print(f"\n(c) Most popular name across all years: {most_popular_name}")
```

End of Problem 3
